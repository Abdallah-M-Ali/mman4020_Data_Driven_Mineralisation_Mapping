```python
import os
import glob
import numpy as np
import pandas as pd

xyz_files = os.path.join(os.path.dirname(os.getcwd()), 'data', "*.xyz")
data_dir = os.path.join(os.path.dirname(os.getcwd()), 'data')
```

```python
df = None
for xyz_file in xyz_files: df = pd.read_table(xyz_file, skiprows=2, delim_whitespace=True,names=['EASTING', 'NORTHING', xyz_file.split("\\")[-1][:-4]])
df
```

```python
mineralogy = pd.read_csv(os.path.join(xyz_dir, "Mineralogy.csv"))
points = list(zip(list(mineralogy["NORTHING_GDA2020"], mineralogy["EASTING_GDA2020"])))
for pt in points: index = df.loc[(df['EASTING']-pt[0]>=0) & (df['EASTING']-pt[0]<250) & (pt[1]-df['NORTHING']>=0) & (pt[1]-df['NORTHING']<250)].index.tolist()
df.to_csv(os.path.join(data_dir, "GawlerTensor.csv"), index=False)
df.loc[df['Mineralogy'] == 1]
```
