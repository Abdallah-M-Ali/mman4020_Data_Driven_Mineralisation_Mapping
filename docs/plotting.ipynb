```python

#import libraries
import cv2
from matplotlib import pyplot as plt
  
# create figure
fig = plt.figure(figsize=(10, 7))
  
# setting values to rows and column variables
rows = 2
columns = 2
  
# reading images
Image1 = cv2.imread('./salliteRGB.png')
Image2 = cv2.imread('./vegitationSatellite.png')
Image3 = cv2.imread('./sateOtherMine.png')
Image4 = cv2.imread('./correctOrientationSatelite.jpg')
  
Image1 = cv2.cvtColor(Image1, cv2.COLOR_BGR2RGB)
Image2 = cv2.cvtColor(Image2, cv2.COLOR_BGR2RGB)
Image3 = cv2.cvtColor(Image3, cv2.COLOR_BGR2RGB)
Image4 = cv2.cvtColor(Image4, cv2.COLOR_BGR2RGB)

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
fig.suptitle("Analysis of a Landsat-8 Imaging Sample in the Gawler Craton", fontsize=16)

# showing image
plt.imshow(Image1)
plt.axis('off')
plt.title("Natural Color Composite using R,G,B Bands")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(Image2, )
plt.axis('off')
plt.title("Normalized Difference Vegetation Index (NDVI)")
  
# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(Image3)
plt.axis('off')
plt.title("Reflectance Distance to Hydrothermal-Mineral Bands")
  
# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(Image4)
plt.axis('off')
plt.title("Reflectance Distance to Iron-Oxide Bands")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

# figure(figsize=(20, 12), dpi=80)
plt.rcParams["figure.figsize"] = [25, 27]
plt.rcParams["figure.autolayout"] = True
def plot_image_grid(images, ncols=None, cmap='gray'):
    '''Plot a grid of images'''
    if not ncols:
        factors = [i for i in range(1, len(images)+1) if len(images) % i == 0]
        ncols = factors[len(factors) // 2] if len(factors) else len(images) // 4 + 1
    nrows = int(len(images) / ncols) + int(len(images) % ncols)
    imgs = [images[i] if len(images) > i else None for i in range(nrows * ncols)]
    f, axes = plt.subplots(nrows, ncols, figsize=(3*ncols, 2*nrows))
    axes = axes.flatten()[:len(imgs)]
    for img, ax in zip(imgs, axes.flatten()):
        ax.axes.yaxis.set_ticks([])
        ax.axes.xaxis.set_ticks([])
        if np.any(img):
            if len(img.shape) > 2 and img.shape[2] == 1:
                img = img.squeeze()
            ax.imshow(img, cmap=cmap)

# make 16 images with 60 height, 80 width, 3 color channels
img1 = mpimg.imread('./salliteRGB.png')
img2 = mpimg.imread('./vegitationSatellite.png')
img3 = mpimg.imread('./sateOtherMine.png')
img4 = mpimg.imread('./correctOrientationSatelite.jpg')

# plot them
plot_image_grid([img1, img2, img3, img4])
```

```python
import numpy as np
from matplotlib import pyplot as plt, cm
from matplotlib.pyplot import figure
plt.rcParams.update({'font.size': 7})
img = mpimg.imread('./drillholeQuality.png')
from matplotlib.ticker import FuncFormatter

figure(figsize=(10, 6), dpi=80)
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.axis('off')
plt.title("Drillhole Locations with Mapped Depths to Crystalline Basement")
colorBlue = (71/255, 118/255, 162/255, 1.0)
colorRed = (255/255, 23/255, 17/255, 1.0)
colorYel = (255/255, 245/255, 61/255, 1.0)
colorLine = (224/255, 175/255, 34/255, 1.0)
plt.plot(0, 0, "o", color=colorBlue, label="All Drillholes", linewidth=6.0)
plt.plot(0, 0, "o", color=colorRed, label="Crystaline Basement Drillholes (CBD)", linewidth=6.0)
plt.plot(0, 0, "o", color=colorYel, label="Randomised 15% of CBD", linewidth=6.0)
plt.plot(0, 0, "-", color=colorLine, label="Depth to Crystalline Basement Contour Lines", linewidth=6.0)

def fmt(x, pos):
    if (pos == 0): return '30--'
    elif (pos == 1): return '70'
    elif (pos == 2): return '110'
    elif (pos ==3): return '150'
    elif (pos == 4): return '190'
    elif (pos == 5): return '230'
    elif (pos == 6): return '270'
    elif (pos == 7): return '310'
    elif (pos == 8): return '350'
    elif (pos == 9): return '400++'
    
x = np.random.rand(100)
y = np.random.rand(100)
z = np.sin(x)+np.cos(y)
contour2 = plt.tricontourf(x, y, z, 100, cmap="gray")
cbar = plt.colorbar(contour2, format=FuncFormatter(fmt))
cbar.set_label('Depths To Crystalline Basement (m)')

plt.imshow(img)
plt.legend(loc="upper left", bbox_to_anchor=(0.01, 1.0))

# plt.show()
# plt.savefig('./QualDrillholeMapping.jpg')
plt.savefig('myimage.png', format='png', dpi=120)
```
